#### Data Dictionary

* Survived: 0 = No, 1 = Yes
* pclass: Ticket class, 1 = 1st, 2 = 2nd, 3 = 3rd
* sibsp: # of siblings / spouses aboard the Titanic
* parch: # of parents / children aboard the Titanic
* ticket: Ticket number
* cabin: Cabin number
* embarked: Port of Embarkation, C = Cherbourg, Q = Queenstown, S = Southampton

### Histograms

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

##### Survived

```python
yes = 0
no = 0
for o in df['Survived']:
    if o == 0:
        no += 1
    else:
        yes += 1
```

```python
plt.bar(['Yes', 'No'], [yes, no], color=['green', 'red'])
plt.xlabel('Survived')
plt.ylabel('# of people')
plt.title('Survivors bar chart')
```

Discovery: About of 350 people survived the titanic, this is around 39.3%.

### Ticket Class

```python
first_class = 0
second_class = 0
thrid_class = 0

for c in df['Pclass']:
    if c == 1:
        first_class += 1
    elif c == 2:
        first_class += 1
    elif c == 3:
        thrid_class += 1
```

```python
plt.pie([first_class, second_class, thrid_class], labels=['1st', '2nd', '3rd'], colors=['#79b7db', '#457996', '#1d3f52'], autopct = '%1.1f%%')
plt.title('Ticket class piechart')
```

Discovery: The 3rd class tickets was most of the population boarding the titanic.

#### Sex
Where 1 = female & 0 = male

```python
counts = df['Sex'].value_counts()
counts
```

```python
male = 0
female = 0

for s in df['Sex']:
    if s == 'male':
        male += 1
    else:
        female += 1
```

```python
plt.bar(['female', 'male'], [female, male], color=['pink','skyblue'])
plt.title('Sex bar chart')
plt.ylabel('# of people')
```

Discovery: Theres almost as twice the male population than the female.

#### Age

```python
df['Age'].plot(kind='hist', color = '#cc5a84', edgecolor = '#732642')
plt.xlabel('Age')
plt.ylabel('# of people')
plt.title('Age histogram')
```

Discovery:Theres way more people around 20-30 years old, and theres almost none 80 years old people.

#### # parents or children

```python
df['Parch'].plot(kind='hist', color= '#769178', edgecolor= '#3d4f3e')
plt.title('Parents/childrens histogram')
plt.xlabel('# of parents/children')
plt.ylabel('# of people')
```

Discovery: Almost 77% of the population have no children.

##### # of siblings or spouses

```python
df['SibSp'].plot(kind = 'hist', color= '#3d4f3e', edgecolor= '#769178')
plt.title('Siblings/spouses histogram')
plt.xlabel('# of siblings/spouses')
plt.ylabel('# of people')
```

Discovery: Around 67.4% of the population had no spouses or siblings.

#### Embarked
Where 2 = Cherbourg, 1 = Queenstown & 0 = Southampton

```python
df['Embarked'] = df['Embarked'].map({'S': 0, 'Q': 1, 'C': 2}) #.map transforms data in a df
df['Embarked'].plot(kind='hist', color='#9a85ab', edgecolor='#6a5978')
plt.title('Embarked histogram')
plt.xlabel('Port')
plt.ylabel('# of people')
```

Discovery:The southhampton port was where most people boarded.

### Class Comparison

#### Survived-Sex

```python
survived_gender = df.groupby(['Survived', 'Sex']).size().unstack()

# Crear el grÃ¡fico de barras apiladas
survived_gender.plot(kind='bar', stacked=True, color=['pink', 'skyblue'])

# Etiquetas y tÃ­tulo
plt.xlabel('Survived')
plt.ylabel('# of people')
plt.title('Survival Count by Gender')
```

Discovery: More woman survived

#### Survived-Class

```python
stacked_data = df.groupby(['Survived', 'Pclass']).size().unstack()


stacked_data.plot(kind='bar', stacked=True, color =['#79b7db', '#457996', '#1d3f52'])

plt.xlabel('Survived')
plt.ylabel('# of people')
plt.title('Survival Count by Passenger Class')
```

Discovery: Almost all people from 3rd class didnt survive

#### Survived-Age

```python
age_range = [0, 10, 20, 30, 40, 50, 60, 70, 80] #We are creating intervalos de 10 en 10
labels = ['0-10', '11-20', '21-30', '31-40', '41-50', '51-60', '61-70', '71-80'] #These are the labels for de x axis using the intervals
df['Age_Group'] = pd.cut(df['Age'], bins=age_range, labels=labels, right=False) #We created column age groups so we can make a bar chart

stacked_age = df.groupby(['Survived', 'Age_Group']).size().unstack()

# Crear el grÃ¡fico de barras apiladas
stacked_age.plot(kind='bar', stacked=False)

plt.xlabel('Survived')
plt.ylabel('# of people')
plt.title('Survival Count by Age Group')
```

Discovery: All the people between 71-80 didnt survive
