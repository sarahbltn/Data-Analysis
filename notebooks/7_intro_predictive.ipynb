```python
import pandas as pd
from sklearn.datasets import make_regression
```

```python
x, y = make_regression(n_features= 1, noise= 30, random_state=0 ) #make regression tiene una correlacion casi perfecta, usamos noise para tener datos mas 'reales'
#Random state sirve para establecer una semilla y poder tener los mismos datos de otra persona
df = pd.DataFrame(x)
df.columns = ['x']
df['y'] = y
df
```

```python
df['x'].plot(kind='hist')
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#Modelo lineal, linea que pase por la mayor cant de puntos posibles
model_1 = LinearRegression() 
model_2 = LinearRegression() 
model_3 = LinearRegression() 

model_1.fit(x,y) #El fit es para entrenar el modelo

```

```python
y_pred = model_1.predict(x)
```

* Una clase es lo que define el comportamiento de un modelo
* Un objeto es la instancia de una clase
* La palabra entrenamiento es meterle datos a tu modelo, le doy x y me tiene que decir la y y asi predecir datos

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
y_mean = np.array([y.mean()] * 100) 
y.mean()
```

```python
##Selector
plt.figure(figsize=(10,3)) #Define el tamano

#Data
plt.scatter(x,y, label= 'Random Data') #Define las variables para hacer la grafica y gracias al legend() se ve el label
#
plt.plot(x,y_mean, label= 'Plot', c='green') #Plot sirve para trazar una linea por todos los puntos en el orden en el que estan
plt.plot(x,y_pred, label= 'Plot', c='red')
##Show
plt.xlabel('Features') #Nombra el eje de las x
plt.ylabel('Target (y)') #Nombra el eje de las y
plt.title('Scatter Example Plot') #Pone un titulo a la grafica
plt.legend()

plt.savefig('../reports/figures/random1.svg') #guarda la imagen donde le digamos
```
